### Box plots

For Price

```python
data["Price"].plot(kind='box',vert=False,title="Price")
```

```python
Q1=np.percentile(data["Price"],25)
Q3=np.percentile(data["Price"],75)
IQR=Q3-Q1
UWL=Q3+(1.5*IQR)
LWL=Q1-(1.5*IQR)
outliers=data[(data["Price"]>UWL) | (data["Price"]<LWL)]
print("Outliers in Price: ",len(outliers))
print("Outlier percentage: ",(len(outliers)/len(data))*100)
```

For Area

```python
data["Area"].plot(kind='box',vert=False,title="Area")
```

```python
Q1=np.percentile(data["Area"],25)
Q3=np.percentile(data["Area"],75)
IQR=Q3-Q1
UWL=Q3+(1.5*IQR)
LWL=Q1-(1.5*IQR)
outliers=data[(data["Area"]>UWL) | (data["Area"]<LWL)]
print("Outliers in Area: ",len(outliers))
print("Outlier percentage: ",(len(outliers)/len(data))*100)
```

For no of bedrooms

```python
data["No. of Bedrooms"].plot(kind='box',vert=False,title="Area")
```

```python
Q1=np.percentile(data["No. of Bedrooms"],25)
Q3=np.percentile(data["No. of Bedrooms"],75)
IQR=Q3-Q1
UWL=Q3+(1.5*IQR)
LWL=Q1-(1.5*IQR)
outliers=data[(data["No. of Bedrooms"]>UWL) | (data["No. of Bedrooms"]<LWL)]
print("Outliers in No. of Bedrooms: ",len(outliers))
print("Outlier percentage: ",(len(outliers)/len(data))*100)
```

after calculating outliers we see

* for price 2825 - 8.5% of our data have outliers
* for area - 1832 - 5.5% of our data have outliers

we are having high outliers in them
because of larger price and area values


so we will apply log tranformation to compress them to compress the large values

* for no of bedrooms - 264 - 0.8% of our data have outliers

we are having ver low outliers for no of bedrooms the outliers were 7,8,9 sharing bedrooms which are rare so lets replace them with 6 share bedrooms

---
### Transformations for Price,Area

```python
data["Price_log"]=np.log1p(data["Price"])
data["Area_log"]=np.log1p(data["Area"])
```

### Repalcing 7,8,9 sharing bedrooms with 6


```python
data["No. of Bedrooms"]=data["No. of Bedrooms"].apply(lambda x: 6 if x > 6 else x)
```

---
### After transformation and replacing lets check results


```python
data["Price_log"].hist()
print("Skewness of price:",data["Price_log"].skew())
```

```python
data["Area_log"].hist()
print("Skewness of area:",data["Area_log"].skew())
```

```python
data["No. of Bedrooms"].hist()
print("Skewness of no of bedrooms:",data["No. of Bedrooms"].skew())
```

Originally, price and area were highly skewed due to very large outliers

After applying log tranformation, the skewness gradually reduced to symmetric now which is normal

for bedrooms , instead of removing the data we replaced them with 6 which is reasonable for housing data

```python
data["Price_log"].plot(kind='box',vert=False,title="Price")
```

## Correlation

```python
numerical_columns = ['Price_log', 'Area_log', 'No. of Bedrooms']
corr = data[numerical_columns].corr()
corr
```

* price_log and area_log has moderate positive correlation
* area_log and no of bedrooms have strong positive correlation
* price_log and no of bedrooms have weak correalation
